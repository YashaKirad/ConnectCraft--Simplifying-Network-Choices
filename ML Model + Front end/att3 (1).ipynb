```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from mlxtend.regressor import StackingRegressor
```

```python
df = pd.read_csv('att3.csv')
df
```

```python
print(df['networkProvider'].unique()) 
```

```python
df.isnull().sum()
```

Bandwidth (in bps) is the maximum rate at which data can be transmitted on a network connection.
Download speed (also in bps) is the actual rate at which data is being received by your device from the network.
Since download speed is typically measured in bytes per second (Bps), while bandwidth is measured in bits per second (bps), you need to convert between the two units. There are 8 bits in a byte. So, to convert from bytes to bits, you multiply by 8.

```python
df['bandwidth'] = df['download'] * 8
df
```

```python

input_latitude = 18.571069

input_longitude = 73.773285
```

```python
# Calculate pairwise distances between input and dataset coordinates
from scipy.spatial.distance import cdist
input_coords = [[input_latitude, input_longitude]]
dataset_coords = df[['location_latitude', 'location_longitude']].values
distances = cdist(input_coords, dataset_coords, metric='euclidean')
```

```python
# Find rows with coordinates similar to input based on 70% similarity
threshold_distance = 0.7 * distances.max()
similar_rows_indices = [i for i, distance in enumerate(distances[0]) if distance <= threshold_distance]
similar_rows = df.iloc[similar_rows_indices]
```

```python
# Print the selected similar rows
print("Rows with similar coordinates to input:")
print(similar_rows)
```

```python
# Define features and target
features = ['download_no_zeros', 'location_latitude', 'location_longitude', 'ping_no_zeros', 'upload_no_zeros']
target = 'bandwidth'

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(similar_rows[features], similar_rows[target], test_size=0.2, random_state=42)
```

```python
# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Initialize models
model1 = RandomForestRegressor()
model2 = GradientBoostingRegressor()
model3 = SVR()
stacking_model = StackingRegressor(regressors=[model1, model2, model3], meta_regressor=Ridge())

# Fit stacking model
stacking_model.fit(X_train_scaled, y_train)
```

```python
# Sample input
sample_input = pd.DataFrame({
    'download_no_zeros': [0],
    'location_latitude': [input_latitude],
    'location_longitude': [input_longitude],
    'ping_no_zeros': [0],       
    'upload_no_zeros': [0]      
})

# Scale sample input
sample_input_scaled = scaler.transform(sample_input)
```

```python
# Predict bandwidth for sample input
sample_prediction = stacking_model.predict(sample_input_scaled)
#print(f"Predicted bandwidth for latitude={input_latitude}, longitude={input_longitude}: {sample_prediction[0]}")
```

```python
# Initialize variables for best provider and best prediction
best_provider = None
best_prediction = float('-inf')

# Iterate over unique network providers in similar rows
for provider in similar_rows['networkProvider'].unique():
    provider_data = similar_rows[similar_rows['networkProvider'] == provider]
    X_provider = provider_data[features]
    X_provider_scaled = scaler.transform(X_provider)

    provider_predictions = stacking_model.predict(X_provider_scaled)
    mean_prediction = provider_predictions.mean()
    
    if mean_prediction > best_prediction:
        best_provider = provider
        best_prediction = mean_prediction
print("Baner -")
print(f"The best provider for latitude={input_latitude}, longitude={input_longitude} is: {best_provider}")
```
